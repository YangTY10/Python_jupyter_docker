```python
from sklearn.datasets import load_boston
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option("display.max_rows", 15)
```

```python
boston = load_boston()
boston
```

```python
df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df["target"] = boston["target"]
df
```

```python
plt.figure(figsize=(14,10))
# y=æ¨™é¡Œè·Ÿåœ–çš„è·é›¢, size=å­—é«”å¤§å°
plt.title("Pearson Correlation of Features", y=1.05, size=15)
# cmap=åœ–æ¯æ ¼é¡è‰², linewidths=æ ¼å­é–“çš„å¯¬åº¦, square=True å°‡æ ¼å­è¨­ç‚ºæ­£æ–¹å½¢, linecolor=æ ¼å­é–“ç·šçš„é¡è‰², annot=True é¡¯ç¤ºç›¸é—œä¿‚æ•¸
sns.heatmap(df.astype(float).corr(), cmap="seismic", linewidths=0.1, square=True, linecolor="white", annot=True)
# å¯æŒ‘é¸å–œæ­¡é¡è‰²æ”¾å…¥cmapåƒæ•¸
# https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(boston["data"], boston["target"], test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(max_depth=3)
regr.fit(x_train, y_train)
```

```python
import numpy as np
predict = regr.predict(x_test)
print("å¯¦éš›çš„åƒ¹éŒ¢:", y_test)
print("é æ¸¬çš„åƒ¹éŒ¢:", predict)
interval = np.subtract(predict, y_test)
print("å·®ç•°:", interval)
```

```python
from sklearn.metrics import r2_score
print("è¨“ç·´è³‡æ–™ r2 score:", r2_score(y_train,regr.predict(x_train)))
print("æ¸¬è©¦è³‡æ–™ r2 score:", r2_score(y_test,regr.predict(x_test)))
```

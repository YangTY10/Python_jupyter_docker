```python
from sklearn.datasets import load_boston
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option("display.max_rows", 15)
```

```python
boston = load_boston()
boston
```

```python
df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df["target"] = boston["target"]
df
```

```python
plt.figure(figsize=(14,10))
# y=標題跟圖的距離, size=字體大小
plt.title("Pearson Correlation of Features", y=1.05, size=15)
# cmap=圖每格顏色, linewidths=格子間的寬度, square=True 將格子設為正方形, linecolor=格子間線的顏色, annot=True 顯示相關係數
sns.heatmap(df.astype(float).corr(), cmap="seismic", linewidths=0.1, square=True, linecolor="white", annot=True)
# 可挑選喜歡顏色放入cmap參數
# https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(boston["data"], boston["target"], test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(max_depth=3)
regr.fit(x_train, y_train)
```

```python
import numpy as np
predict = regr.predict(x_test)
print("實際的價錢:", y_test)
print("預測的價錢:", predict)
interval = np.subtract(predict, y_test)
print("差異:", interval)
```

```python
from sklearn.metrics import r2_score
print("訓練資料 r2 score:", r2_score(y_train,regr.predict(x_train)))
print("測試資料 r2 score:", r2_score(y_test,regr.predict(x_test)))
```
